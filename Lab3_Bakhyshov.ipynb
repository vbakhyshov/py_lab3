```python
import pandas as pd
import re

# Приклад створення DataFrame
data = {
    'Country': ["Republic of Korea", "United States of America", "United Kingdom of Great Britain and Northern Ireland", "China, Hong Kong Special Administrative Region", "Bolivia (Plurinational State of)", "Switzerland17"],
    'Energy Supply': [100, 200, 150, 120, 80, 90]
}

energy_df = pd.DataFrame(data)

# Функція для видалення цифр та дужок з назв країн
def clean_country_name(country):
    # Використання регулярного виразу для видалення цифр та дужок
    return re.sub(r'[\d\(\)]', '', country)

# Застосування функції до стовпця 'Country'
energy_df['Country'] = energy_df['Country'].apply(clean_country_name)

# Виведення тільки для обраних країн
selected_countries = ['American Samoa', 'South Korea', 'Bolivia']
selected_energy_data = energy_df.loc[energy_df['Country'].isin(selected_countries)]

# Виведення тільки для обраних країн з використанням loc
print(selected_energy_data)

# Виведення тільки для обраних країн з використанням loc із змінною energy_df
print(energy_df.loc[energy_df['Country'].isin(['American Samoa', 'South Korea', 'Bolivia'])])
```

```python
def task_eleven():
    # Створення стовпця з чисельністю населення
    Result['Population Estimate'] = Result['Energy Supply'] / Result['Energy Supply per Capita']
    
    # Сортування за чисельністю населення та вибірка шостої країни, якщо вона існує
    sorted_result = Result.sort_values(by='Population Estimate', ascending=False)
    
    # Визначення шостої країни та населення
    sixth_country = sorted_result.iloc[5] if len(sorted_result) >= 6 else None
    
    # Повернення кортежу із назвою країни та населенням
    return (sixth_country.name, sixth_country['Population Estimate']) if sixth_country is not None else None

# Виклик функції
result_tuple = task_eleven()

# Перевірка, чи є результат
if result_tuple is not None:
    print(result_tuple)
else:
    print("Недостатньо даних для оцінки чисельності населення для шостої країни.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Задання стилів seaborn для кращого вигляду графіку
sns.set()

# Створення бульбашкової діаграми
plt.figure(figsize=(12, 8))
bubble_plot = sns.scatterplot(x='Rank', y='% Renewable', size='2015', hue='Continent', data=Result, sizes=(10, 1000),
                              alpha=0.7, palette='viridis', edgecolor='w', linewidth=0.5)

# Додавання підписів для кожної бульбашки (назва країни)
for line in range(0, Result.shape[0]):
    bubble_plot.text(Result['Rank'].iloc[line], Result['% Renewable'].iloc[line], Result.index[line],
                     horizontalalignment='left', size='small', color='black')

# Додавання заголовка та підписів до вісей
plt.title('Dependency of % Renewable on Rank (Bubble Size: GDP in 2015)', fontsize=16)
plt.xlabel('Rank', fontsize=14)
plt.ylabel('% Renewable', fontsize=14)

# Видалення легенди
bubble_plot.get_legend().remove()

# Показ графіку
plt.show()
```
